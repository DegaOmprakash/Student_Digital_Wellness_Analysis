```python
# Final "Conclusion" Chart
plt.figure(figsize=(8, 8))
df['Affects_Academic_Performance'].value_counts().plot.pie(
    autopct='%1.1f%%', 
    startangle=90, 
    colors=['#ff9999','#66b3ff'], 
    explode=(0.05, 0), 
    pctdistance=0.85
)

# Create the "Donut" hole
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title("Overall Impact: Does Social Media Affect Academic Performance?")
plt.ylabel('') # Hides the 'Affects_Academic_Performance' label on the side
plt.show()
```

```python
sns.histplot(x=df['Age'],y=df['Avg_Daily_Usage_Hours'],bins=5)
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.select_dtypes(include=['float64', 'int64']).corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
```

```python
# Final Summary Analysis
correlations = df.select_dtypes(include=['float64', 'int64']).corr()['Addicted_Score'].sort_values(ascending=False)

print("--- FINAL PROJECT SUMMARY ---")
print(f"1. Strongest Negative  Correlation: Mental_Health_Score ({correlations['Mental_Health_Score']:.2f})")
print(f"2. Usage vs. Addiction: {correlations['Avg_Daily_Usage_Hours']:.2f}")
print(f"3. Impact on Sleep (Negative Correlation): {correlations['Sleep_Hours_Per_Night']:.2f}")
print("-" * 30)
print("CONCLUSION: High social media usage is a primary driver for sleep deprivation ")
print("and shows a near-linear relationship with mental health challenges and social conflicts.")
```
